````python
import  twitter,io,json
````

````python
def load_json(filename):
    with io.open('resources/{0}.json'.format(filename),'r', encoding='utf-8') as f: # hier muss r stehen, damit die txt-Datei gelesen werden kann
        readfile = f.read()
        jsonfile = json.loads(readfile)
        return jsonfile
````

2\. Gib die ersten 10 Tweets, Usernamen und Hashtags aus der Datei aus.

````python
status_texts = [ status['text'] for status in trumptweets ] #Tweets
    
usernames = [ user_mention['screen_name'] for status in trumptweets 
                                        for user_mention in status['entities']['user_mentions'] ] #User

hashtags = [ hashtag['text'] for status in trumptweets
                                       for hashtag in status['entities']['hashtags']] #Hashtags

print(json.dumps(status_texts[0:10], indent=1))
print(json.dumps(usernames[0:10], indent=1))
print(json.dumps(hashtags[0:10], indent=1))
````

3\. Erstelle eine Sammlung aller WÃ¶rter, die in den Tweets vorkommen und gib die ersten 30 WÃ¶rter aus.

````python
words = [ w 
             for statuses in status_texts
                for w in statuses.split() ]

print(json.dumps(words[:30], indent=1))
````

4\. Erstelle nun eine HÃ¤ufigkeitsverteilung der Elemente WÃ¶rter, Usernamen und Hashtags. Lass dir die hÃ¤ufigsten 10 Ergebnisse in einer Tabelle anzeigen.

````python
from collections import Counter
from prettytable import PrettyTable


for label, data in (('Word', words),
                    ('Screen Name', usernames),
                    ('Hashtag', hashtags)):
    pt = PrettyTable(field_names=[label, 'Count'])
    
    c = Counter(data)
    [ pt.add_row(kv) for kv in c.most_common()[:10] ]
    pt.align[label], pt.align['Count'] = 'l', 'r' # Spaltenausrichtung festlegen
    print(pt)
````

````python
retweets = [
            # Speichern eines Tupels dieser 3 Werte...
            (status['retweet_count'],
             status['retweeted_status']['user']['screen_name'],
             status['text'])
            
            #... fÃ¼r jeden Status...
            for status in trumptweets
    
            #... sofern der Status diese Bedingung erfÃ¼llt
                if 'retweeted_status' in status
            ]

#Ausgabe der ersten 5 des sortierten Ergebnisses
pt = PrettyTable(field_names=['Count', 'Screen Name', 'Text'])
#Darstellung jedes Elemens im Tupel
[ pt.add_row(row) for row in sorted(retweets, reverse=True)[:5] ] 
pt.max_width['Text'] = 50
pt.align = 'l'
print(pt)
````

6\. Erstelle eine Graphik, die dir anzeigt, wie hÃ¤ufig die am meist geteilten Tweets von Donald Trump geteilt wurden. 

````python
import matplotlib.pyplot as plt



counts = [count for count, _, _ in retweets]

plt.hist(counts)
plt.title("Retweets")
plt.xlabel("Bins (Anzahl von Retweets)")
plt.ylabel("Anzahl von Tweets in einem bin")
plt.show()
````
